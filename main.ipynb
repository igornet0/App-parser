```python
coins = {}

for coin in data_manager.coin_list:
    for path in data_manager.get_path(data_type="raw",
                                      coin=coin,
                                      timetravel="5m"):
        coins.setdefault(coin, [])
        coins[coin].append(path)
```

```python
from backend.Dataset import DatasetTimeseries
```

```python
coins

```

```python
for coin, paths in coins.items():
    path = data_manager.create_dir("processed", coin)
    data_manager.create_dir("processed", path / "5m")
    paths_pd = list(map(DataManager.get_df, paths))
    dt = DatasetTimeseries.concat_dataset(*paths_pd)
    dt = DatasetTimeseries(dt)
    dt_new = dt.clear_dataset()
    dt.set_dataset(dt_new)
    # print(coin, dt.get_dataset().head())
    # dt.set_path_save(path / "5m")
    # dt.save_dataset(f"clear-{coin}-5m.csv")
    coins[coin] = dt
    
```
